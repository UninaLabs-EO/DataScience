```python
from skimage import exposure, io
import numpy as np
import matplotlib.pyplot as plt
from osgeo import gdal
from pathlib import Path
from glob import glob
import time, os
gdal.UseExceptions()

def convert(img_array, output: str, path, mode='std'):
     os.makedirs(output, exist_ok=True)
     if mode == 'std':
          d=img_array.std()
          m=img_array.mean()
          scale = 3.5
          a = int(m - scale*d)
          b = int(m + scale*d)
     else:
          cdf, bins = exposure.cumulative_distribution(img_array, nbins=2**16)
          a = np.where(cdf > 0.0001)[0][0]
          b = np.where(cdf > 0.9999)[0][0]
     
     img_array_clip = np.clip(img_array, a_min=a, a_max=b)
     img_eq = exposure.equalize_adapthist(img_array_clip, clip_limit=0.08)
     img_eq *= 255
     img_eq = img_eq.astype('uint8')
     io.imsave(output+'/'+ path.stem+'.png', img_eq)
```

```python
def convertAllBand(folder:str, b: int):
     files = glob('DATA/*.tif')
     for f in files:
          path = Path(f)
          ds = gdal.Open(path.as_posix(), gdal.GA_ReadOnly)
          rb = ds.GetRasterBand(b)
          try:
               img_array = rb.ReadAsArray()
               convert(img_array,output=folder, path=path)
               # time.sleep(1)
          except RuntimeError:
               print('Error file not working')
```

```python
convertAllBand(folder='MultiSpectral/scale35/CL008/B2', b=1)
convertAllBand(folder='MultiSpectral/scale35/CL008/B3', b=2)
convertAllBand(folder='MultiSpectral/scale35/CL008/B4', b=3)
convertAllBand(folder='MultiSpectral/scale35/CL008/B8', b=4)
```
